### Import dependencies

```python
from langsmith import Client
```

### Attach Feedback to a specific trace in LangSmith

019c3b73-cf15-7e10-838f-8e2f6c28d049

```python
client = Client()
```

```python
RUN_ID = "019c3b73-cf15-7e10-838f-8e2f6c28d049"
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source="api",
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments",
    value="This is a long form free text feedback left by Wilfried",
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments with score",
    value="This is a long form free text feedback left by Wilfried",
    score=0,
    feedback_source_type="api"
)
```
